```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(100)
infected = 0.3 * np.exp(0.1 * time)
beds_tot = 1000
hospitalized = np.zeros(100)
input = np.zeros(100)
output = np.zeros(100)
for t in range(1, 100):
    beds = beds_tot - hospitalized[t-1]
    input[t] = min(infected[t], beds)
    output[t] = 0 if t < 7 else input[t-7]
    dn = input[t] - output[t]
    hospitalized[t] = hospitalized[t-1] + dn
```

```python
plt.scatter(time, infected)
plt.scatter(time, hospitalized)
```

```python
import pandas as pd

df = pd.DataFrame({'infected': infected, 'hospitalized': hospitalized})
```
